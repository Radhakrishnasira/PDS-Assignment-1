```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
clean_data=pd.read_csv('Frailtycleandata.csv')
```

```python
clean_data
```

```python
plt.hist(clean_data['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
correlation_matrix =clean_data[['Height(in)', 'Weight(lb)', 'Age', 'Grip_strength(kg)']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(data=clean_data, vars=['Height(in)', 'Weight(lb)', 'Age', 'Grip_strength(kg)'], hue='Frailty', palette='Set1')
plt.title('Pairplot of Numeric Variables by Frailty')
plt.show()
```
