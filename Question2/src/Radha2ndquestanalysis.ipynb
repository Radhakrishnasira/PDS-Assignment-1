```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('clean_data_studentperformance.csv')
```

```python
 data
```

```python
colors = ["yellow", "blue"]  # Blue and Orange

# Create countplot with specified colors
sns.countplot(data=data, x='gender', palette=colors)
plt.title('Distribution of Gender')
plt.show()
```

```python
colors = ["yellow", "blue"]  # Blue and Orange
sns.boxplot(data=data, x='gender', y='Total Score',palette=colors)
plt.title('Total Score Distribution by Gender')
plt.show()
```

```python
#colors=["orange","green","Yellow"]
sns.scatterplot(data=data, x='test preparation course', y='Total Score',palette='Set2')
plt.title('Total Score Distribution by preparation course')
plt.show()
```

```python
sns.barplot(data=data, x='race/ethnicity', y='Total Score',palette='Set1')
plt.title('Total Score Distribution by preparation course')
plt.show()
```

```python
sns.pairplot(data=data, vars=['math score', 'reading score', 'writing score', 'Total Score'], hue='parental level of education',palette='Set1')
plt.title('Pairplot of Scores')
plt.show()
```
